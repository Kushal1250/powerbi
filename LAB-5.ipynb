```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
sns.set_theme(style="darkgrid")
sns.regplot(x="total_bill", y="tip", data=tips,  x_jitter=True, ci=50, marker="o", line_kws={"color": "red", "linewidth": 2})
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip Regression Plot with Customizations')
plt.savefig('Exp1')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.lmplot(data=titanic,x='age', y='fare', col='sex', hue='class',
           height=5, aspect=0.7, scatter_kws={"s": 50}, 
           line_kws={"linewidth": 2}, order=2)
# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare Regression Plot by Sex and Class')
plt.savefig('Exp2')
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="coolwarm",  annot=True, fmt=".1f", linewidths=0.5,linecolor='black', cbar_kws={'label': 'Number of Passengers'} )

plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.title("Heatmap of Number of Passengers per Month Across Different Years")

# plt.show()
plt.savefig('Exp3')
```

```python
iris = sns.load_dataset("iris")
iris_data = iris.drop(columns=["species"])
correlation_matrix = iris_data.corr()

sns.clustermap(correlation_matrix,method="ward",metric="euclidean",cmap=sns.color_palette("coolwarm", as_cmap=True),row_cluster=False,annot=True,fmt=".2f",)

plt.title("Clustered Heatmap of Iris Dataset Correlation Matrix")
plt.show()
# plt.savefig('Exp4')
```

```python
iris = sns.load_dataset("iris")

g = sns.pairplot(iris,hue="species", diag_kind="kde", markers=["o", "s", "D"], plot_kws={'alpha': 0.7})

g.fig.suptitle("Pairwise Relationships in the Iris Dataset", y=1.02)
plt.savefig('Exp5')
```

```python
tips = sns.load_dataset("tips")

g = sns.jointplot(
    data=tips, 
    x="total_bill", 
    y="tip", 
    kind="scatter", 
    cmap="coolwarm", 
    marginal_kws=dict(bins=30), 
    height=8
)
g.fig.suptitle("Joint Plot of Total Bill vs Tip", y=1.03)
plt.show()
plt.savefig('Exp6')
```

```python
titanic = sns.load_dataset("titanic")

g = sns.FacetGrid(titanic, col="class", row="sex",  despine=False)
g.map(sns.histplot, "age", kde=True)

g.set_titles("{col_name} - {row_name}")
g.fig.suptitle("Age Distribution by Class and Sex", y=1.05)

plt.show()
plt.savefig('Exp7')
```

```python
tips = sns.load_dataset("tips")

g = sns.catplot(data=tips, x="day",y="total_bill", hue="sex", kind="violin", dodge=True,alpha=0.7)

g.set_xticklabels(rotation=45)
g.fig.suptitle("Comparison of Day vs Total Bill for Different Times")

plt.show()
plt.savefig('Exp8')
```

```python
diamonds = pd.read_csv('Desktop/diamonds.csv')

sns.lmplot(data=diamonds, x="carat", y="price", hue="cut", palette="muted", ci=50,)

plt.yscale("log")
plt.title("Regression Plot of Carat vs Price Faceted by Cut")

plt.show()
plt.savefig('Exp9')
```

```python
sns.set_style("whitegrid")
flights = sns.load_dataset("flights")

# For scatter plot
sns.relplot(
    data=flights, 
    x="year", 
    y="passengers", 
    hue="month", 
    style="month", 
    kind="scatter", 
    height=6, 
    aspect=1.5
)

plt.title("Scatter Plot of Year vs Passengers Faceted by Month")
plt.show()
plt.savefig()
```
