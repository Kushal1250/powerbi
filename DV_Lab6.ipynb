```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
print(tips.head(),"\n")
flights = sns.load_dataset('flights')
print(flights.head(),"\n")
titanic = sns.load_dataset('titanic')
print(titanic.head(),"\n")
iris = sns.load_dataset('iris')
print(iris.head(),"\n")
```

```python
#Question 1
sns.lmplot(data=tips,x='total_bill',y='tip',hue='smoker',markers=['o','h'],col="time")
# plt.legend()
```

```python
#Question 2
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
corr_matrix = flights_pivot.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, linewidths=0.5, fmt=".2f")
```

```python
titanic_filtered = titanic[["pclass", "sex", "age"]].dropna()
titanic_filtered["sex"] = titanic_filtered["sex"].map({"male": 0, "female": 1})
sns.clustermap(titanic_filtered, method="ward", cmap='coolwarm',standard_scale=1)
plt.show()
```

```python
# 4. Compare numerical variables in the Iris dataset
sns.pairplot(iris, hue="species", kind="reg", palette="Set2")
plt.show()
```

```python
# 5. Create a joint plot for total bill vs. tip amount
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex", alpha=0.5)
plt.show()
```

```python
# 6. Compare fare prices across passenger classes using categorical plots
sns.catplot(data=titanic, x="class", y="fare", kind="box", hue="sex")
sns.swarmplot(data=titanic, x="class", y="fare", hue="sex", dodge=True)
plt.show()
```

```python
# 7. Analyze the effect of age on survival rate using a relational plot
sns.relplot(data=titanic, x="age", y="survived", hue="survived", size="survived", alpha=0.6)
plt.show()
```

```python
# 8. Compare petal length vs. petal width in the Iris dataset using lmplot
sns.lmplot(data=iris, x="petal_length", y="petal_width", hue="species", markers=["o", "s", "^"], height=6, aspect=1)
plt.show()
```

```python
# 9. Create a FacetGrid for monthly flight data
fg = sns.FacetGrid(flights, col="year", col_wrap=4)
fg.map(sns.lineplot, "month", "passengers")
plt.show()
```

```python
# 10. Analyze survival across passenger classes using multi-plot comparisons
sns.relplot(data=titanic, x="class", y="survived", hue="survived", size="age", col="embark_town", style="sex")
plt.show()
```
